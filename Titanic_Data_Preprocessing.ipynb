# Dataset Description
The Titanic dataset contains information about passengers aboard the Titanic.
The objective is to explore and preprocess the data in order to prepare a clean dataset
for further analysis or machine learning.

Main features include:
- Passenger class
- Sex
- Age
- Fare
- Embarkation port

The target variable is **Survived**, which indicates whether a passenger survived or not.

```python
# for missign values
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.drop('Cabin', axis=1, inplace=True, errors='ignore')
```

```python
df.drop(['Name', 'Ticket', 'PassengerId'], axis=1, inplace=True)
```

```python
# male=0 & female=1
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# Embarked: One-hot encoding
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```

```python
columns_to_drop = ['Name', 'Ticket', 'PassengerId', 'Cabin']
existing_columns_to_drop = [col for col in columns_to_drop if col in df.columns]

df.drop(existing_columns_to_drop, axis=1, inplace=True)
```

```python
# Final check
print("\nMissing values after preprocessing:\n", df.isnull().sum())
print("\nData types:\n", df.dtypes)
print("\nFirst 5 rows:\n", df.head())
```
